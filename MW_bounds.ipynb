```python
import matplotlib.pyplot as plt
import numpy             as np

plt.rc('text', usetex=True)
```

```python
def convex_lower_bound(x, beta=0.5):
    return beta**x

def convex_upper_bound(x, beta=0.5):
    return 1 - (1 - beta) * x
```

```python
x = np.arange(-1.0, 2.0, 0.001)

plt.figure(dpi=800)
#plt.title(r"$\forall x \in [0, 1],\ \beta^x \leq u_\beta(x) \leq 1 - (1 - \beta)x$")

plt.plot(x, convex_lower_bound(x, 0.5), label=r"$\beta^x$")
plt.plot(x, convex_upper_bound(x, 0.5), label=r"$1 - (1 - \beta)x$")

plt.legend(loc="upper left")
plt.show()
```

```python
def convex_lower_bound_eta(x, eta=0.5):
    return (1 - eta)**x

def convex_upper_bound_eta(x, eta=0.5):
    return (1 - eta * x)
```

```python
x = np.arange(-1.0, 2.0, 0.001)

plt.figure(dpi=800)
#plt.title(r"$\forall x \in [0, 1],\ (1 - \eta x) \leq u_\eta(x) \leq (1 - \eta)^x$")

plt.plot(x, convex_lower_bound_eta(x, 0.5), label=r"$(1 - \eta)^x$")
plt.plot(x, convex_upper_bound_eta(x, 0.5), label=r"$(1 - \eta x)$")

plt.legend(loc="upper left")
plt.show()
```

```python
def one_minus(x):
    return 1 - x

def one_minus_upper(x):
    return np.exp(-1 * x)

def one_minus_lower(x):
    return np.exp(-1 * x - x**2)
```

```python
x = np.arange(-1.0, 2.0, 0.001)

plt.figure(dpi=800)
#plt.title(r"$\forall x \in [0, \frac{1}{2}],\ 1 - x \leq \exp(-x)$")

plt.plot(x, one_minus(x),       label=r"$1 - x$")
plt.plot(x, one_minus_upper(x), label=r"$\exp(-x)$")


plt.legend(loc="upper left")
plt.show()
```

```python
x = np.arange(-1.0, 2.0, 0.001)

plt.figure(dpi=800)
#plt.title(r"$\forall x \in [0, \frac{1}{2}],\ \exp(-x - x^2) \leq 1 - x$")

plt.plot(x, one_minus(x),       label=r"$1 - x$")
plt.plot(x, one_minus_lower(x), label=r"$\exp(-x - x^2)$")


plt.legend(loc="upper left")
plt.show()
```
